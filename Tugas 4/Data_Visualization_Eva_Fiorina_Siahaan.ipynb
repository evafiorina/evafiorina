ChatGPT Prompt: Beri kode untuk menampilkan boxplot dari fitur-fitur numerik



```python
# Menampilkan boxplot untuk setiap fitur numerik
numerical_features = ['carat', 'depth', 'table', 'price', 'x', 'y', 'z']
for feature in numerical_features:
    plt.figure(figsize=(10, 6))
    sns.set_palette("Blues")
    sns.boxplot(x=df[feature])
    plt.title('Boxplot of Numeric Features ' + feature)
    plt.show()
```

ChatGPT Prompt: beri kode untuk menampilkan heat map dari fitu-fitur numerik

```python
# @title Heat Map
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Membaca file CSV dan membuat DataFrame
df = pd.read_csv('/content/drive/MyDrive/Dataset/diamonds.csv')

# Memilih hanya fitur numerik
numeric_features = df.select_dtypes(include=['float64', 'int64'])

# Menghitung korelasi antar fitur numerik
correlation_matrix = numeric_features.corr()

# Membuat heat map
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap Korelasi Fitur Numerik')
plt.show()
```
