```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Membaca dataset
startup_data = pd.read_csv('/content/drive/MyDrive/Dataset/50_Startups.csv')  # Ganti 'nama_file.csv' dengan nama file dataset sesuai dengan lokasi Anda

# Memilih hanya kolom-kolom numerik
numerical_columns = startup_data.select_dtypes(include=['int', 'float'])

# Membuat boxplot untuk setiap fitur numerik
plt.figure(figsize=(12, 8))
sns.boxplot(data=numerical_columns)
plt.title('Boxplot untuk Fitur Numerik')
plt.xticks(rotation=45)  # Untuk memutar label sumbu x agar lebih mudah dibaca
plt.show()
```

```python
# @title
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Membaca file CSV ke dalam DataFrame
df = pd.read_csv('/content/drive/MyDrive/Dataset/50_Startups.csv')

# Memilih hanya kolom numerik
numeric_columns = df.select_dtypes(include=['int64', 'float64'])

# Menentukan ukuran plot
plt.figure(figsize=(14, 10))

# Melakukan loop untuk setiap fitur numerik dan membuat violin plot
for i, col in enumerate(numeric_columns.columns):
    plt.subplot(3, 3, i + 1)
    sns.violinplot(data=df, y=col)
    plt.title(col)

plt.tight_layout()
plt.show()
```

```python
# @title Heat Map
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Membaca file CSV dan membuat DataFrame
df = pd.read_csv('/content/drive/MyDrive/Dataset/50_Startups.csv')

# Memilih hanya fitur numerik
numeric_features = df.select_dtypes(include=['float64', 'int64'])

# Menghitung korelasi antar fitur numerik
correlation_matrix = numeric_features.corr()

# Membuat heat map
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap Korelasi Fitur Numerik')
plt.show()
```
